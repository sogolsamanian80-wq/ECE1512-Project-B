```python
# New plots

import re
import matplotlib.pyplot as plt


# ---------- READ FILE ----------
fname = "Focal_1_4.txt"
with open(fname, "r") as f:
    txt = f.read()

# ---------- REGEX FOR EPOCH METRICS ----------
epoch_pattern = re.compile(
    r"\* Acc@1 ([0-9\.]+) loss ([0-9\.]+) auroc ([0-9\.]+) f1_score ([0-9\.]+)"
)

# ---------- REGEX FOR BEST-EPOCH BLOCK ----------
best_pattern = re.compile(
    r"Results on best epoch:\s*\{([^}]+)\}",
    re.MULTILINE
)

# ---------- COLLECT ALL EXPERIMENTS ----------
experiments = []  # each element: dict with epoch metrics + best metrics

# Split by "Used config:" because each block starts with that
blocks = txt.split("Used config:")
blocks = blocks[1:]  # first chunk before the first config is irrelevant

for block in blocks:
    # Parse epoch-level metrics
    accs, losses, aucs, f1s = [], [], [], []

    for acc, loss, auc, f1 in epoch_pattern.findall(block):
        accs.append(float(acc))
        losses.append(float(loss))
        aucs.append(float(auc))
        f1s.append(float(f1))

    # Parse the best-epoch summary
    best = best_pattern.search(block)
    best_dict = {}
    if best:
        raw = best.group(1).strip()
        for kv in raw.split(","):
            if ":" in kv:
                key, val = kv.split(":")
                key = key.strip().strip("'")
                val = val.strip()
                try:
                    val = float(val)
                except ValueError:
                    pass
                best_dict[key] = val

    experiments.append({
        "acc_vector": accs,
        "loss_vector": losses,
        "auc_vector": aucs,
        "f1_vector": f1s,
        "best": best_dict,
    })

acc1=experiments[0]["acc_vector"]
loss1=experiments[0]["loss_vector"]
AUC1=experiments[0]["auc_vector"]
F11=experiments[0]["f1_vector"]

val_acc1 = acc1[0::2]
val_loss1 = loss1[0::2]
val_auc1 = AUC1[0::2]
val_f1 = F11[0::2]

test_acc1  = acc1[1::2]
test_loss1  = loss1[1::2]
test_auc1  = AUC1[1::2]
test_f1= F11[1::2]

# Plotting
epochs = list(range(1, len(val_acc1)+1))

# -------------------------------  Camelyon16 -------------------------------------------------------
labels = ['Validation','Test']
fig = plt.figure(figsize=(12,8))
fig.suptitle('Camelyon16 Focal Loss (alpha = 1, gamma = 4)')

# Accuracy
plt.subplot(2,2,1)
plt.plot(epochs, val_acc1, linestyle='-', color='blue')
plt.plot(epochs, test_acc1, linestyle='--', color='blue')

plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend(labels)

# Loss
plt.subplot(2,2,2)
plt.plot(epochs, val_loss1, linestyle='-', color='red')
plt.plot(epochs, test_loss1, linestyle='--', color='red')

plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy Loss')
plt.grid(True)
plt.legend(labels)

# AUROC
plt.subplot(2,2,3)
plt.plot(epochs, val_auc1, linestyle='-', color='green')
plt.plot(epochs, test_auc1, linestyle='--', color='green')

plt.title('AUROC over Epochs')
plt.xlabel('Epoch')
plt.ylabel('AUROC')
plt.grid(True)
plt.legend(labels)

# F1-score
plt.subplot(2,2,4)
plt.plot(epochs, val_f1, linestyle='-', color='purple')
plt.plot(epochs, test_f1, linestyle='--', color='purple')

plt.title('F1-score over Epochs')
plt.xlabel('Epoch')
plt.ylabel('F1-score')
plt.grid(True)
plt.legend(labels)

plt.tight_layout()
plt.show()
# -------------------------------  Camelyon17 -------------------------------------------------------

acc1=experiments[1]["acc_vector"]
loss1=experiments[1]["loss_vector"]
AUC1=experiments[1]["auc_vector"]
F11=experiments[1]["f1_vector"]

val_acc1 = acc1[0::2]
val_loss1 = loss1[0::2]
val_auc1 = AUC1[0::2]
val_f1 = F11[0::2]

test_acc1  = acc1[1::2]
test_loss1  = loss1[1::2]
test_auc1  = AUC1[1::2]
test_f1= F11[1::2]

epochs = list(range(1, len(test_acc1)+1))

# Plotting
fig = plt.figure(figsize=(12,8))
fig.suptitle('Camelyon17 Focal Loss (alpha = 1, gamma = 4)')

# Accuracy
plt.subplot(2,2,1)
plt.plot(epochs, val_acc1, linestyle='-', color='blue')
plt.plot(epochs, test_acc1, linestyle='--', color='blue')

plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend(labels)

# Loss
plt.subplot(2,2,2)
plt.plot(epochs, val_loss1, linestyle='-', color='red')
plt.plot(epochs, test_loss1, linestyle='--', color='red')

plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy Loss')
plt.grid(True)
plt.legend(labels)

# AUROC
plt.subplot(2,2,3)
plt.plot(epochs, val_auc1, linestyle='-', color='green')
plt.plot(epochs, test_auc1, linestyle='--', color='green')

plt.title('AUROC over Epochs')
plt.xlabel('Epoch')
plt.ylabel('AUROC')
plt.grid(True)
plt.legend(labels)

# F1-score
plt.subplot(2,2,4)
plt.plot(epochs, val_f1, linestyle='-', color='purple')
plt.plot(epochs, test_f1, linestyle='--', color='purple')

plt.title('F1-score over Epochs')
plt.xlabel('Epoch')
plt.ylabel('F1-score')
plt.grid(True)
plt.legend(labels)

plt.tight_layout()
plt.show()

# -------------------------------  BARCS DATASET -------------------------------------------------------
acc1=experiments[2]["acc_vector"]
loss1=experiments[2]["loss_vector"]
AUC1=experiments[2]["auc_vector"]
F11=experiments[2]["f1_vector"]

val_acc1 = acc1[0::2]
val_loss1 = loss1[0::2]
val_auc1 = AUC1[0::2]
val_f1 = F11[0::2]

test_acc1  = acc1[1::2]
test_loss1  = loss1[1::2]
test_auc1  = AUC1[1::2]
test_f1= F11[1::2]
epochs = list(range(1, len(val_acc1)+1))

# Plotting
fig = plt.figure(figsize=(12,8))
fig.suptitle('BRACS Focal Loss (alpha = 1, gamma = 4)')

# Accuracy
plt.subplot(2,2,1)
plt.plot(epochs, val_acc1, linestyle='-', color='blue')
plt.plot(epochs, test_acc1, linestyle='--', color='blue')

plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.legend(labels)

# Loss
plt.subplot(2,2,2)
plt.plot(epochs, val_loss1, linestyle='-', color='red')
plt.plot(epochs, test_loss1, linestyle='--', color='red')

plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Cross-Entropy Loss')
plt.grid(True)
plt.legend(labels)

# AUROC
plt.subplot(2,2,3)
plt.plot(epochs, val_auc1, linestyle='-', color='green')
plt.plot(epochs, test_auc1, linestyle='--', color='green')

plt.title('AUROC over Epochs')
plt.xlabel('Epoch')
plt.ylabel('AUROC')
plt.grid(True)
plt.legend(labels)

# F1-score
plt.subplot(2,2,4)
plt.plot(epochs, val_f1, linestyle='-', color='purple')
plt.plot(epochs, test_f1, linestyle='--', color='purple')

plt.title('F1-score over Epochs')
plt.xlabel('Epoch')
plt.ylabel('F1-score')
plt.grid(True)
plt.legend(labels)

plt.tight_layout()
plt.show()
```
